```python
# first step: Processing Data
# Categorical figures identification
# column are not yet categorical
#exploring categorical features
print('Training set:')
for col_name in df.columns:
  if df[col_name].dtypes == 'object':
    unique_cat = len(df[col_name].unique())
    print("Feature '{col_name}'has {unique_cat} categories".format(col_name=col_name, unique_cat=unique_cat))

# observe how distributed the feature service is, it is distributed evenly, we make  dummies for them
print()
print('Distribution of service categories:')
print(df['service'].value_counts().sort_values(ascending=False).head())
```

```python
# Label Encoder
#Inserting category features into 2D numpy array

from sklearn.preprocessing import LabelEncoder,OneHotEncoder
categorical_columns=['protocol_type', 'service', 'flag']
#to get list of categorical columns into variable, categorical_columns inset code
categorical_columns=['protocol_type', 'service', 'flag']
#Get categorical values into a 2d numpy array
df_categorical_values=df[categorical_columns]
testdf_categorical_values=df_test[categorical_columns]
df_categorical_values.head()

```

```python
# making dummies column names
#protocol typt
unique_protocol=sorted(df.protocol_type.unique())
string1='Protocol_type_'
unique_protocol2=[string1 + x for x in unique_protocol]\
#services
unique_service=sorted(df.service.unique())
string2='service_'
unique_service2=[string2 + x for x in unique_service]
# flag
unique_flag=sorted(df.flag.unique())
string3='flag_'
unique_flag2=[string3 + x for x in unique_flag]
#put together
dumcols=unique_protocol=unique_protocol2 + unique_service2 + unique_flag2
print(dumcols)

# Do likwise for test set
unique_service_test=sorted(df_test.service.unique())
unique_service2_test=[string2 + x for x in unique_service_test]
testdumcols=unique_protocol2 + unique_service2_test + unique_flag2
```

```python
## Using LabelEncoders Transform categorical features into numbers
df_categorical_values_enc=df_categorical_values.apply(LabelEncoder().fit_transform)
print(df_categorical_values_enc.head())
# test set
testdf_categorical_values_enc=testdf_categorical_values.apply(LabelEncoder().fit_transform)
```

```python
## One-hot-Encoding
enc=OneHotEncoder()
df_categorical_values_encenc=enc.fit_transform(df_categorical_values_enc)
df_cat_data=pd.DataFrame(df_categorical_values_encenc.toarray(),columns=dumcols)
# test set
testdf_categorical_values_encenc=enc.fit_transform(testdf_categorical_values_enc)
testdf_cat_data=pd.DataFrame(testdf_categorical_values_encenc.toarray(),columns=testdumcols)

df_cat_data.head()
```

```python
# Adding 6 missing categories from train set to test set
trainservice=df['service'].tolist()
testservice=df_test['service'].tolist()
difference=list(set(trainservice) - set(testservice))
string='service_'
difference=[string + x for x in difference]
difference
```

```python
for col in difference:
  testdf_cat_data[col]=0

testdf_cat_data.shape

```

```python
#Join encoded categorical dataframe with non categorical dataframe
newdf=df.join(df_cat_data)
newdf.drop('flag', axis=1, inplace=True)
newdf.drop('protocol_type', axis=1, inplace=True)
newdf.drop('service', axis=1, inplace=True)
#test data
newdf_test=df_test.join(testdf_cat_data)
newdf_test.drop('flag', axis=1, inplace=True)
newdf_test.drop('protocol_type', axis=1, inplace=True)
newdf_test.drop('service', axis=1, inplace=True)
print(newdf.shape)
print(newdf_test.shape)
```
