3.3.3 CODE to build a linear regression model

Building a linear regression model
Now you have created your feature and target arrays, you will train a linear regression model on all feature and target values.

As the goal is to assess the relationship between the feature and target values there is no need to split the data into training and test sets.

X and y have been preloaded for you as follows:

X = [[600], [800], [1100], [1400], [1800]]

y = [[700], [900], [1300], [1600], [1900]]

Linear regression is a fundamental statistical and machine learning technique used for modeling the relationship between a dependent variable (target) and one or more independent variables (features). Its primary goal is to find the best-fitting linear relationship

**2. Fill the gaps with code to build, fit, and predict price using a linear regression model.**

```python
from sklearn.linear_model import LinearRegression
# Training data rent (y) based on square feet (X)
X = [[600], [800], [1100], [1400], [1800]]
y = [[700], [900], [1300], [1600], [1900]]
# Create the Linear Regression model & # Fill in the blank with 'LinearRegression()'
model = LinearRegression()
# Fit the model using Training data & # Fill in the blank with writing the 'model'
model.fit(X, y)
# Predict the rent of a apt sized 1200 square feet, so method for predictiopn can be used for getting apt rent price
print ('The rent for 1200 should be : $%.2f' % model.predict([[1200]])[0])
```

In Python, you can implement Support Vector Machines (SVM) using libraries like scikit-learn. Import the SVM class, prepare your dataset, create an SVM instance with desired parameters, use the .fit() method to train the model, and then make predictions using .predict(). SVMs are efficient and well-supported in Python for various applications.

```python
# Import the train_test_split function from scikit-learn to split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve, classification_report, precision_recall_curve
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GridSearchCV, RandomizedSearchCV
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['sms_spam_svm.csv']))
# Dataset is now stored in a Pandas Dataframe
y = df.iloc[:, 0].values
y = np.where(y == 'spam', -1, 1)
X = df.iloc[:, [1, 2]].values

# test_size=0.3 indicates that 30% of the data will be used for testing, random_state=0 for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Create an instance of the Support Vector Machine (SVM) classifier with specific parameters.
svm = SVC(kernel='linear', C=1.0, random_state=1, probability=True)
```

**3. Fill in the training data in fit () as arguments based on the slides.**

```python
# Fit (train) the SVM classifier on the training data
svm.fit(X_train, y_train)
```

**4. Fill the predict method to obtain class labels.**

```python
# Make predictions on the test data using the SVM classifier
y_pred = svm.predict(X_test)
```

**5. Fill the gaps with code in predict_proba().**

```python
# Fit (train) the SVM classifier on the training data
svm.fit(X_train , y_train)  # Fill in the training data as arguments

# Make predictions on the test data using the SVM classifier
y_pred = svm.predict(X_test)  # Use the predict method to obtain class labels

# Calculate the confusion matrix to evaluate the classification performance
cm = confusion_matrix(y_test, y_pred)

# Predict class probabilities on the test data using the SVM classifier
predicted_probab_log = svm.predict_proba(X_test)

# Extract the predicted probabilities for the positive class (usually class 1)
predicted_probab_log = predicted_probab_log[:, 1]
```

**6. Fill the blank in the following code.**

```python
# Calculate the False Positive Rate (FPR) and True Positive Rate (TPR) for Receiver Operating Characteristic (ROC) curve
FPR, TPR , _= roc_curve(y_test, predicted_probab_log)  #
```
